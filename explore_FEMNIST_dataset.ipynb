```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_dir = "datasets\FEMNIST_by_write\write_all_labels.hdf5"
binary_data_file = h5py.File(dataset_dir, "r")

writers = sorted(binary_data_file.keys())
```

```python
img_count_by_wirter = []
count_covered_class = [0]*63

for writer in writers:
    images = binary_data_file[writer]['images']
    labels = binary_data_file[writer]['labels']
    img_count_by_wirter.append(images.shape[0])
    count_covered_class[len(np.unique(labels))] += 1
```

```python
img_count_by_wirter[3500:4000]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(range(len(count_covered_class)), count_covered_class, color='r')
ax.set_xlabel('Number of covered classes')
ax.set_ylabel('Number of writers')
ax.set_title('Writers covered classes')

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# ax.hist(img_count_by_wirter, bins=100, color='r')
ax.hist(img_count_by_wirter, bins=100, rwidth=0.8, color='green')
ax.set_xlabel('Number of images')
ax.set_ylabel('Number of writers')
ax.set_title('Writers images')

fig.show()
```
