```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_dir = "datasets\FEMNIST_by_write\write_all_labels.hdf5"
binary_data_file = h5py.File(dataset_dir, "r")

writers = sorted(binary_data_file.keys())
```

```python
class_cover_count = [0]*63
for writer in writers:
    images = binary_data_file[writer]['images']
    labels = binary_data_file[writer]['labels']
    class_cover_count[len(np.unique(labels))] += 1
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(range(len(class_cover_count)), class_cover_count, color='r')
ax.set_xlabel('Number of covered classes')
ax.set_ylabel('Number of writers')
ax.set_title('Writers covered classes')

fig.show()
```
